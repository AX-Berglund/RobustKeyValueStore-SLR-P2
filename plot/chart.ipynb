```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    "N": [3, 3, 3, 10, 10, 10, 100, 100, 100],
    "M": [3, 10, 100, 3, 10, 100, 3, 10, 100],
    "PUT Latency (ms)": [1.5, 1.2, 0.42, 1.72, 1.08, 0.18, 0.76, 0.62, 0.59],
    "GET Latency (ms)": [0.67, 0.9, 0.14, 2.06, 0.53, 0.15, 0.48, 0.54, 0.48]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = np.arange(len(df))

bars1 = ax.bar(index - bar_width/2, df["PUT Latency (ms)"], bar_width, label="PUT Latency (ms)", color="orange")
bars2 = ax.bar(index + bar_width/2, df["GET Latency (ms)"], bar_width, label="GET Latency (ms)", color="blue")

for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

ax.set_xlabel("Configuration (N, M)")
ax.set_ylabel("Latency (ms)")
ax.set_title("PUT and GET Latency for Different N and M Configurations")
ax.set_xticks(index)
ax.set_xticklabels([f"({n}, {m})" for n, m in zip(df["N"], df["M"])], rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data_new = {
    "M": [100, 150, 200],
    "PUT Latency (ms)": [0.59, 0.6, 0.67],
    "GET Latency (ms)": [0.48, 0.42, 0.46]
}

df_new = pd.DataFrame(data_new)

fig, ax = plt.subplots(figsize=(8, 6))

bar_width = 0.4
index = range(len(df_new))

bars1 = ax.bar(index, df_new["PUT Latency (ms)"], bar_width, label="PUT Latency (ms)", color="orange")
bars2 = ax.bar([i + bar_width for i in index], df_new["GET Latency (ms)"], bar_width, label="GET Latency (ms)", color="blue")

for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

ax.set_xlabel("Number of Operations (M)")
ax.set_ylabel("Latency (ms)")
ax.set_title("PUT and GET Latency for N=100 with Varying M")
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(df_new["M"])
ax.legend()

plt.tight_layout()
plt.show()
```
